```python
df.drop("Column1", axis = 1, inplace = True)
```

```python
df = df.rename(columns={"Cost":"Total Cost"})
```

```python
df["Unit Price"] = (df["Unit Price"]).round(2)
df["Profit"] = (df["Unit Price"] - df["Unit Cost"]).round(2)
df["Total Profit"] = df["Quantity"] * df["Profit"]
df["Profit margin"] = (df["Total Profit"] / df["Revenue"]).round(2)
```

```python
df.head()
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(df.corr(), annot = True, cmap = "coolwarm", fmt =".2f")
plt.show()
```

```python
# Convert Date to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract year and month
df['year_month'] = df['Date'].dt.to_period('M')

monthly_profit = df.groupby('year_month')['Profit'].sum()
monthly_revenue = df.groupby('year_month')['Revenue'].sum()
monthly_profit_margin = df.groupby('year_month')['Profit margin'].mean()
monthly_quantity = df.groupby('year_month')['Quantity'].sum()


# Create a figure with 1 row and 2 columns for side-by-side plots
fig, axes = plt.subplots(2, 2, figsize=(12, 5))  # 1 row, 2 columns

# Plot Profit per Month
axes[0, 0].plot(monthly_profit.index.astype(str), monthly_profit.values, marker='o', linestyle='-', color='b')
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Total Profit')
axes[0, 0].set_title('Profit Per Month')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].grid(True)

# Plot Revenue per Month (Top Right)
axes[0, 1].plot(monthly_revenue.index.astype(str), monthly_revenue.values, marker='o', linestyle='-', color='g')
axes[0, 1].set_xlabel('Month')
axes[0, 1].set_ylabel('Total Revenue')
axes[0, 1].set_title('Revenue Per Month')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].grid(True)

# Plot Profit Margin Per Month (Bottom Left)
axes[1, 0].plot(monthly_profit_margin.index.astype(str), monthly_profit_margin.values, marker='o', linestyle='-', color='r')
axes[1, 0].set_xlabel('Month')
axes[1, 0].set_ylabel('Profit Margin')
axes[1, 0].set_title('Profit Margin Per Month')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True)

# Plot Unit Price Per Month (Bottom Right)
axes[1, 1].plot(monthly_quantity.index.astype(str), monthly_quantity.values, marker='o', linestyle='-', color='m')
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Sale quantity')
axes[1, 1].set_title('Sale quantity Per Month')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(True)


# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
# Group by Year-Month and Sub Category, summing the profit
country_month_profit = df.groupby(['year_month', 'Sub Category'])['Total Profit'].sum().unstack()

# Calculate the percentage change (change rate) per month
country_month_profit_pct_change = country_month_profit.pct_change().fillna(0)  # fillna(0) to avoid NaN in the first month

# Scale the y-axis for better visualization of large changes
fig, ax = plt.subplots(figsize=(14, 7))
country_month_profit_pct_change.plot(kind='bar', stacked=True, colormap='tab20', ax=ax)

# Labels and title
ax.set_xlabel('Month-Year')
ax.set_ylabel('Percentage Change in Profit')
ax.set_title('Stacked Bar Chart: Percentage Change of Profit by Product Category per Month')
ax.tick_params(axis='x', rotation=45)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Optionally, scale the y-axis to better show large changes
ax.set_ylim([-10, 10])  # You can adjust this range as needed

# Show the plot
plt.show()
```
